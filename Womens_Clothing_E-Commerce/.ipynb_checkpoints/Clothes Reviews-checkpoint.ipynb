* Tokens of the reviews

```python
def extract_tokens(words):
    words = re.sub("[^a-zA-Z]"," ", words)
    text = words.lower().split()                   
    return " ".join(text)
```

```python
data['review'] = data['review'].astype('str')
data['tokens_reviews'] = data['review'].apply(extract_tokens)
```

#### REMOVE STOP WORDS

```python
stop_words = stopwords.words('english')
print(stop_words[::10])
```

```python
#clothing stopwords
clothes =['dress','color','wear','top','sweater','material','shirt','jeans','pant',
          'skirt','order','white','black','fabric','blouse','sleeve','even', 'jacket']
```

```python
def stopwords(words):
    review = [text.lower() for text in words.split() if text.lower() not in 
           stop_words and text.lower() not in clothes]
    return " ".join(review)
```

```python
data['tokens_reviews'] = data['review'].apply(stopwords)
data.head()
```

#### REMOVE PUNCS and DIGITS

```python
# count unnecessary parts
re_punc = r'[^\w\s\d]'
re_digits = r'\d+'

print("Number of digits in reviews ", data['tokens_reviews'].str.count(re_digits).sum())
print("Number of puncs in reviews ", data['tokens_reviews'].str.count(re_punc).sum())
```

```python
def remove_puncs_digits(words):
    #remove digits
    words = re.sub(re_digits, '', words)
    #remove punctuaions
    words = re.sub(re_punc, '', words)
    
    return words

data['tokens_reviews'] = data['tokens_reviews'].apply(lambda x: remove_puncs_digits(x))
```

```python
# count again after cleaning unnecessary parts
re_punc = r'[^\w\s\d]'
re_digits = r'\d+'

print("Number of digits in reviews ", data['tokens_reviews'].str.count(re_digits).sum())
print("Number of puncs in reviews ", data['tokens_reviews'].str.count(re_punc).sum())
```

### MODELS

Let's remove unnecessary column review and then split data to prepare models

```python
reviews.drop('review', axis=1, inplace=True)
reviews = reviews.rename(columns={"tokens_reviews": "review"})
```

```python
X = reviews["review"]
y = reviews["recommended"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=101)
```

#### CountVectorizer

```python
vectorizer = CountVectorizer()

X_train_count = vectorizer.fit_transform(X_train)
X_test_count = vectorizer.transform(X_test)
```

```python
X_train_count.toarray()
```

```python
X_train_count.get_shape()
```

#### Tf-Idf Vectorizer

```python
tf_idf_vectorizer = TfidfVectorizer()

X_train_tf_idf = tf_idf_vectorizer.fit_transform(X_train)
X_test_tf_idf = tf_idf_vectorizer.transform(X_test)
```

```python
X_train_tf_idf.toarray()
```

```python
X_train_tf_idf.get_shape()
```

* MACHINE LEARNING MODELS

##### Logistic Regression with Count Vectorizer

```python
log_model_count = LogisticRegression(max_iter=200, n_jobs=-1)
log_model_count.fit(X_train_count, y_train)
log_pred_count = log_model_count.predict(X_test_count)

log_acc_count = accuracy_score(y_test, log_pred_count.round())
print("Accuracy : ", log_acc_count)
log_f1_count = f1_score(y_test, log_pred_count.round())
print("F1 Score : ", log_f1_count)

ConfusionMatrixDisplay.from_predictions(y_test, log_pred_count, cmap = 'Reds')
```

##### Logistic Regression with Tf-Idf Vectorizer

```python
log_model_tf = LogisticRegression(max_iter=200, n_jobs=-1)
log_model_tf.fit(X_train_tf_idf, y_train)
log_pred_tf = log_model_tf.predict(X_test_tf_idf)

log_acc_tf = accuracy_score(y_test, log_pred_tf.round())
print("Accuracy : ", log_acc_tf)
log_f1_tf = f1_score(y_test, log_pred_tf.round())
print("F1 Score : ", log_f1_tf)

ConfusionMatrixDisplay.from_predictions(y_test, log_pred_tf, cmap = 'Reds')
```

##### Support Vector Machine (SVM) with Count Vectorizer

```python
svm_model_count = svm.SVC(kernel='linear') # Linear Kernel
svm_model_count.fit(X_train_count, y_train) # Train the model using the training sets
svm_pred_count = svm_model_count.predict(X_test_count) # Predict the response for test dataset

svm_acc_count = accuracy_score(y_test, svm_pred_count.round())
print("Accuracy : ", svm_acc_count)
svm_f1_count = f1_score(y_test, svm_pred_count.round())
print("F1 Score : ", svm_f1_count)

ConfusionMatrixDisplay.from_predictions(y_test, svm_pred_count, cmap = 'Reds')
```

##### Support Vector Machine (SVM) with Tf-Idf Vectorizer

```python
svm_model_tf = svm.SVC(kernel='linear') # Linear Kernel
svm_model_tf.fit(X_train_tf_idf, y_train) #Train the model using the training sets
svm_pred_tf = svm_model_tf.predict(X_test_tf_idf) # Predict the response for test dataset

svm_acc_tf = accuracy_score(y_test, svm_pred_tf.round())
print("Accuracy : ", svm_acc_tf)
svm_f1_tf = f1_score(y_test, svm_pred_tf.round())
print("F1 Score : ", svm_f1_tf)

ConfusionMatrixDisplay.from_predictions(y_test, svm_pred_tf, cmap = 'Reds')
```

##### XGBClassifier with Count Vectorizer

```python
xgb_model_count = XGBClassifier(n_estimators=1000, learning_rate=0.05, max_depth = 10, use_label_encoder=False)
xgb_model_count.fit(X_train_count, y_train) # Train the model using the training sets
xgb_pred_count = xgb_model_count.predict(X_test_count) # Predict the response for test dataset

xgb_acc_count = accuracy_score(y_test, xgb_pred_count.round())
print("Accuracy : ", xgb_acc_count)
xgb_f1_count = f1_score(y_test, xgb_pred_count.round())
print("F1 Score : ", xgb_f1_count)

ConfusionMatrixDisplay.from_predictions(y_test, xgb_pred_count, cmap = 'Reds')
```

##### XGBClassifier with Tf-Idf Vectorizer

```python
xgb_model_tf = XGBClassifier(n_estimators=1000, learning_rate=0.05, max_depth = 10, use_label_encoder=False)
xgb_model_tf.fit(X_train_tf_idf, y_train) #Train the model using the training sets
xgb_pred_tf = xgb_model_tf.predict(X_test_tf_idf) # Predict the response for test dataset

xgb_acc_tf = accuracy_score(y_test, xgb_pred_tf.round())
print("Accuracy : ", xgb_acc_tf)
xgb_f1_tf = f1_score(y_test, xgb_pred_tf.round())
print("F1 Score : ", xgb_f1_tf)

ConfusionMatrixDisplay.from_predictions(y_test, xgb_pred_tf, cmap = 'Reds')
```

##### LIGHTGBM with Count Vectorizer

```python
# lightGBM expects float32/64 instead of int32/64. That's why for this model, I created another count-vectorizer.
vectorizer_lgb = CountVectorizer(dtype=np.float32)

X_train_count_lgb = vectorizer_lgb.fit_transform(X_train)
X_test_count_lgb = vectorizer_lgb.transform(X_test)

lgb_model_count = lgb.LGBMClassifier(objective='binary', learning_rate=0.05, n_estimators=1000)
lgb_model_count.fit(X_train_count_lgb, y_train, eval_set = [(X_test_count_lgb, y_test)],
                    early_stopping_rounds=20,
                    verbose=10
                    )
lgb_pred_count = lgb_model_count.predict(X_test_count_lgb, num_iteration=lgb_model_count.best_iteration_)

lgb_acc_count = accuracy_score(y_test, lgb_pred_count.round())
print("Accuracy : ", lgb_acc_count)
lgb_f1_count = f1_score(y_test, lgb_pred_count.round())
print("F1 Score : ", lgb_f1_count)

ConfusionMatrixDisplay.from_predictions(y_test, lgb_pred_count, cmap = 'Reds')
```

##### LIGHTGBM with Tf-Idf Vectorizer

```python
# lightGBM expects float32/64 instead of int32/64. That's why for this model, I created another count-vectorizer.
tf_idf_vectorizer = TfidfVectorizer(dtype=np.float32)

X_train_tf_idf_lgb = tf_idf_vectorizer.fit_transform(X_train)
X_test_tf_idf_lgb = tf_idf_vectorizer.transform(X_test)

lgb_model_tf = lgb.LGBMClassifier(objective='binary', learning_rate=0.05, n_estimators=1000)
lgb_model_tf.fit(X_train_tf_idf_lgb, y_train, eval_set = [(X_test_tf_idf_lgb, y_test)],
                    early_stopping_rounds=20,
                    verbose=10
                    )
lgb_pred_tf = lgb_model_tf.predict(X_test_tf_idf_lgb, num_iteration=lgb_model_tf.best_iteration_)

lgb_acc_tf = accuracy_score(y_test, lgb_pred_tf.round())
print("Accuracy : ", lgb_acc_tf)
lgb_f1_tf = f1_score(y_test, lgb_pred_count.round())
print("F1 Score : ", lgb_f1_tf)

ConfusionMatrixDisplay.from_predictions(y_test, lgb_pred_tf, cmap = 'Reds')
```

### COMPARISON OF THESE MODELS

```python
model_scores_acc = [ log_acc_count, log_acc_tf, 
                     svm_acc_count, svm_acc_tf,
                     xgb_acc_count, xgb_acc_tf, 
                     lgb_acc_count, lgb_acc_tf ]

model_scores_f1 = [ log_f1_count, log_f1_tf, 
                    svm_f1_count, svm_f1_tf,
                    xgb_f1_count, xgb_f1_tf, 
                    lgb_f1_count, lgb_f1_tf ]

model_names = ['Logistic Regression with Count Vectorizer',         'Logistic Regression with Tf-Idf Vectorizer',
               'Support Vector Machine with Count Vectorizer',      'Support Vector Machine with Tf-Idf Vectorizer',
               'XGBClassifier with Count Vectorizer',               'XGBClassifier with Tf-Idf Vectorizer',
               'LIGHTGBM with Count Vectorizer',                    'LIGHTGBM with Tf-Idf Vectorizer' ]

evaluate = pd.DataFrame(columns=["Model","Accuracy-Score", "F1-Score"])

for name, score_acc, score_f1 in zip(model_names, model_scores_acc, model_scores_f1):
    row = evaluate.shape[0]
    evaluate.loc[row,"Model"] = name
    evaluate.loc[row,"Accuracy-Score"] = score_acc.round(5)
    evaluate.loc[row,"F1-Score"] = score_f1.round(5)

evaluate
```

##### Based on Accuracy Score

```python
## Visualization
fig, ax = plt.subplots(figsize=(16,9))

## Encoded dataset
bar = sns.barplot(x = evaluate["Model"], y = evaluate["Accuracy-Score"])
for rec in bar.patches:
    height = rec.get_height()
    ax.text(rec.get_x() + rec.get_width()/2, height*1.02, height,ha="center")
ax.set_title("Accuracy Score")
ax.set_xticklabels(evaluate["Model"].to_list(),rotation =75)
```

```python
## Visualization
fig, ax = plt.subplots(figsize=(16,9))

## Encoded dataset
bar = sns.barplot(x = evaluate["Model"], y = evaluate["F1-Score"])
for rec in bar.patches:
    height = rec.get_height()
    ax.text(rec.get_x() + rec.get_width()/2, height*1.02, height,ha="center")
ax.set_title("Accuracy Score")
ax.set_xticklabels(evaluate["Model"].to_list(),rotation =75)
```
